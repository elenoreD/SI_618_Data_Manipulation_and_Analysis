### <span style="color:magenta"> Q4. [15 points] Calculate the standard deviation and 95% confidence interval (i.e. the range between the 2.5th and 97.5th percentiles) of "fixed acidity".  Use these values to determine to what extent the distribution of "fixed acidity" conforms to the empirical rule (i.e. that 95% of the values fall within 2 standard deviations of the mean). </span>

```python
fixed_acidity_std = np.std(wine['fixed acidity'])
confidence_interval = np.percentile(wine['fixed acidity'],2.5),'-',np.percentile(wine['fixed acidity'],97.5)
print(fixed_acidity_std, confidence_interval)
```

It shows the 95% confidence interval of "fixed acidity" has a range of nearly 7, which equals to roughly 4 standard deviations of the mean. This means the distribution of "fixed acidity" doesn't conform to the empirical rule.
